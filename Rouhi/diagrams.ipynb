## Some diagrams for our hydrogen energy paper

```python
from graphviz import Digraph 
```

```python
# sabatier
f = Digraph('hydrogen production', format='svg', filename='hydrogen.gv', graph_attr = {'splines':'ortho'})
f.attr(rankdir='LR') #, size='10,6')

f.attr('node', shape = 'none')
f.node('grid', label = 'Electricity grid')
# f.node('elec_heat', label = 'Electricity or heat')
# f.node('elec_heat2', label = 'Electricity or heat')
f.attr('node', shape = 'trapezium')
f.node('storage', 'Storage site')

f.attr('node', shape = 'rectangle')
f.node('renewable', label = 'Renewable energy source', fillcolor='green', style='filled')
f.node('gas', label = 'Gas', fillcolor='gray', style='filled')
f.node('oil', label = 'Oil', fillcolor='black', style='filled', fontcolor='white')
f.node('coal', label = 'Coal', fillcolor='saddlebrown', style='filled', fontcolor='white')
f.node('biomass', label = 'Biomass', fillcolor='springgreen', style='filled', fontcolor='black')

f.node('elec', label = 'Electricity')
f.node('elec2', label = 'Electricity')
f.node('elec3', label = 'Electricity')

f.node('water', label = 'Water', fillcolor='aqua', style='filled')
f.node('gasification', label = 'Gasification')
f.node('smr', label = 'SMR')
f.node('h2', label = 'Hydrogen', fontcolor='blue')
f.node('h2g', label = 'Hydrogen', fontcolor='green')
f.node('co2', label = 'Carbon dioxide')
# f.node('fossil', label = 'Fossil fuels \n (gas\oil\coal)')
f.node('electrolysis', label = 'Electrolysis')
# f.node('cells', label = 'Windmills, solar cells, etc')
f.node('pp', label = 'Power plants')

f.edge('renewable', 'elec2')
f.edge('coal', 'pp')
f.edge('oil', 'pp')
f.edge('gas', 'pp')
f.edge('pp', 'elec')
# f.edge('cells', 'elec2')
f.edge('pp', 'co2')

f.edge('gas', 'smr')
f.edge('coal', 'gasification')
f.edge('biomass', 'gasification')

f.edge('smr', 'h2')
f.edge('smr', 'co2')
f.edge('gasification', 'h2')
f.edge('gasification', 'co2')
f.edge('elec2', 'electrolysis')
f.edge('water', 'electrolysis')
f.edge('electrolysis', 'h2g')
f.edge('h2', 'elec3')
f.edge('h2g', 'elec3')

f.edge('elec3', 'grid')
f.edge('elec', 'grid')
f.edge('elec2', 'grid')
f.edge('co2', 'storage')

# f.edge('H2', 'HP', label='Compression')
# f.edge('HP', 'Methane')
# f.edge('Air', 'CD')
# f.edge('CD', 'CO2')
# f.edge('Water', 'Electrolysis')
# f.edge('Electrolysis', 'H2')
# f.edge('Heat', 'HP', color='red')
# f.node('Heat', color='red', fontcolor='red')

f.render()
f
```
