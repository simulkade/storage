# Renewable electricity and a future without fossil fuel
A future energy market without fossil fuel is a societal demand especially in the developed (i.e. rich) countries. Renewable electricity is a volatile energy resource. To control its fluctuation one needs to store it in large scale. Storage comes at a price, i.e. electricity loss. Here, I'm looking at the current electricity production of Denmark, assume a certain increase in capacity, and try to understand what needs to be done for energy storage before we can say goodbye to oil and gas forever.
